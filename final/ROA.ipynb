```python
import pandas as pd
import numpy as np
```

```python
file = pd.read_csv('roa.csv',index_col=0,parse_dates=True)
file_1 = file[['datadate','tic','at','ni']]
file_1.head()
```

```python
with open('chosen_list.txt', 'r') as f:
    List = [line.split('\n')[0] for line in f]
List
```

```python
file_1['ROA'] = file_1['ni']/file_1['at']
```

```python
file_1
```

```python
lst = list(set(file_1['tic']))
cnt = 0
for i in List:
    if i in lst:
        cnt += 1
print(cnt)
```

```python
roa_dic = {}
for idx, row in file_1.iterrows():
    year = int(row[0].split('/')[0])
    tic = row[1]
    roa = float(row[4])
    if tic not in List:
        continue
    if tic not in roa_dic:
        roa_dic[tic] = [0 for _ in range(10)]
    if 2000 <= year <= 2009:
        roa_dic[tic][year - 2000] = roa
```

```python
for tic in roa_dic:
    if 0 in roa_dic[tic] or float('NaN') in roa_dic[tic]:
        print(tic, roa_dic[tic])
```

```python
import pickle
with open('dic.pkl', 'rb') as f:
    dic = pickle.load(f)
```

```python
cnt = 0
for comp in dic:
    if comp not in roa_dic:
        continue
    dic[comp]['ROA'] = []
    for i in range(10):
        dic[comp]['ROA'] += [roa_dic[comp][i] for _ in range(3)]
print(len(dic['AAPL']['ROA']))
```

```python
with open('dic.pkl', 'wb') as f:
    pickle.dump(dic, f)
```
